### Create URL endpoint

```
POST /users
GET /users
GET /organization
```

```
# ไม่เหมาะ
POST /createUser
GET /getAllUsers
GET /getOrganization
```

### Resource Naming Convention

```
//ใช้ plural nouns: (เติม S) 
GET /api/users แทน /api/user 

//ใช้ path parameters สำหรับ specific resource: 
GET /api/users/{id}

//ใช้ query parameters สำหรับ filtering/sorting: 
GET /api/users?role=admin&sort=name
```

### API versioning

```
POST /api/v1/users
POST /api/v2/users
```

### response format

```
{
  status : 200,
  data : {
    user: {
      count: 10,
      data: [...]
    },
    organization : {
      count: 5,
      data: [...]
    }

  }
  meta : {
    page: 1,
    take: 5,
    total: 100
  }
}
```
